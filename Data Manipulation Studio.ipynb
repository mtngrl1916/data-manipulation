## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

baltimore = pd.read_csv('dataset/baltimore_9-24-2016_9-30-2017.csv')
boston = pd.read_csv('dataset/boston_9-24-2016_9-30-2017.csv')
new_york = pd.read_csv('dataset/new-york_9-24-2016_9-30-2017.csv')
philadelphia = pd.read_csv('dataset/philadelphia_9-24-2016_9-30-2017.csv')
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
baltimore_pct_dict = {}
for col in baltimore.columns:
    pct_missing_baltimore = np.mean(baltimore[col].isnull())
    baltimore_pct_dict[col] = '{}%'.format(round(pct_missing_baltimore*100))
baltimore_pct_dict
```

```python
# Clean your data here! BOSTON

boston_pct_dict = {}
for col in boston.columns:
    pct_missing_boston = np.mean(boston[col].isnull())
    boston_pct_dict[col] = '{}%'.format(round(pct_missing_boston*100))
boston_pct_dict
```

```python
# Clean your data here! NEW YORK

new_york_pct_dict = {}
for col in new_york.columns:
    pct_missing_new_york = np.mean(new_york[col].isnull())
    new_york_pct_dict[col] = '{}%'.format(round(pct_missing_new_york*100))
new_york_pct_dict
```

```python
# Clean your data here! PHILADELPHIA

philadelphia_pct_dict = {}
for col in philadelphia.columns:
    pct_missing_philadelphia = np.mean(philadelphia[col].isnull())
    philadelphia_pct_dict[col] = '{}%'.format(round(pct_missing_philadelphia*100))

philadelphia_pct_dict
```

```python
check_columns = pd.DataFrame({'baltimore': baltimore_pct_dict,
                             'boston': boston_pct_dict,
                             'new_york': new_york_pct_dict,
                             'philadelphia': philadelphia_pct_dict})
check_columns
```

```python
cities = [baltimore,boston,new_york,philadelphia]
empty_columns = ['Type','Grade','Environment','Quality','Condition','Appearance','Storage','Crop','Trans Mode']
for city in cities:
    city = city.drop(empty_columns, axis=1,inplace=True)
```

```python
philadelphia.head()
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
all_cities = pd.concat([baltimore,boston,new_york,philadelphia])
all_cities.tail()
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
all_cities["Unit of Sale"].fillna('Undefined',inplace=True)
all_cities.groupby('Unit of Sale').agg({'Low Price':'mean','High Price':'mean'})
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
all_cities.groupby('Variety').size()
```
